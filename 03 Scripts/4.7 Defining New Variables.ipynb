# 03. If-Statements: User-Defined Functions

```python
# Creating subset to avoid memory issues

df = df_ords_prods_merged[:1000000]
```

```python
# Defining function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Using function on df

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# 'price_range' value count

df['price_range'].value_counts(dropna = False)
```

```python
# Maximum price

df['prices'].max()
```

### Q4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Frequency of 'order_hour_of_day' values

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating loop with results in 3 categories

results_3 = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if (value >= 9) and (value <= 16):
        results_3.append("Most orders")
    elif (value == 7) or (value == 8):
        results_3.append("Average orders")
    elif (value >= 17) and (value <= 22):
        results_3.append("Average orders")
    else:
        results_3.append("Fewest orders")
```

```python
# Showing list of 'results_3' values

results_3
```

```python
# Creating new column

df_ords_prods_merged['busiest_period_of_day'] = results_3
```
