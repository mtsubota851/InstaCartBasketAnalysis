### Q4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions

```python
# Aggregating data with transform() function

df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking data

df_ords_prods_merged.head(10)
```

```python
# Adding 'loyalty_flag' column using loc() function
# Above 40 orders = Loyal customer

df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# More than 10, less than or equal to 40 orders = Regular customer

df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Less than or equal to 10 orders = New customer

df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Value counts of 'loyalty_flag'

df_ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

### Q6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria: 

a. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

b. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Creating new column in df for average price of products

df_ords_prods_merged['avg_order_cost_by_cust'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform('mean')
```

```python
# Checking columns

df_ords_prods_merged.head()
```

```python
# Creating spender flag: low spender - less than 10

df_ords_prods_merged.loc[df_ords_prods_merged['avg_order_cost_by_cust'] < 10, 'spender_flag'] = 'Low spender'
```

```python
# Creating spender flag: High spender - higher than or equal to 10

df_ords_prods_merged.loc[df_ords_prods_merged['avg_order_cost_by_cust'] >= 10, 'spender_flag'] = 'High spender'
```

```python
# Checking columns

df_ords_prods_merged.head(10)
```

### Q7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

a. If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

b. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

c. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create new column for median of 'days_since_prior_order'

df_ords_prods_merged['order_frequency_by_cust'] = df_ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Checking columns

df_ords_prods_merged.head()
```

```python
# Creating order frequency flag: Non-frequent - higher than 20

df_ords_prods_merged.loc[df_ords_prods_merged['order_frequency_by_cust'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Creating order frequency flag: Regular customer - more than 10, lower than or equal to 20

df_ords_prods_merged.loc[(df_ords_prods_merged['order_frequency_by_cust'] <= 20) & (df_ords_prods_merged['order_frequency_by_cust'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
# Creating order frequency flag: Frequent customer - lower than or equal to 10

df_ords_prods_merged.loc[df_ords_prods_merged['order_frequency_by_cust'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
